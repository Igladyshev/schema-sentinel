```python
from dataclasses import dataclass


@dataclass
class ClassA:
    a: str
    b: str

    def __eq__(self, other):
        if not isinstance(other, ClassA):
            raise Exception("Not Implemented")
        for attribute, value in self.__dict__.items():
            other_val = getattr(other, attribute)
            print(f"other.{attribute}={other_val}, self.{attribute}={value}")
            if other_val != value:
                return False
        return True


a = ClassA("one", "two")
b = ClassA("one", "three")

a == b
```

```python
cls = ClassA
isinstance(b, cls)
```
